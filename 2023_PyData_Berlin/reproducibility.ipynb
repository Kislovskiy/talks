```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
from pathlib import Path
```

```python
Path().resolve().parent/"results"
```

```python
fig, ax = plt.subplots()
function = "sin"
x = np.linspace(0, 10, 100)
ax.plot(x, getattr(np, function)(x))
ax.set_title(f"{function}")
plt.show()
fig.savefig(f"results/{function}.pdf", format='pdf', bbox_inches='tight', pad_inches=0.5)
```
