```python
import pandas as pd
from matplotlib import pyplot as plt
from pathlib import Path
```

```python
def plot_global_temp(ax):
    global_temp_path = (
        Path(".").resolve().parent.parent.joinpath("data", "global_temperature.txt")
    )

    pd.read_csv(
        global_temp_path,
        sep=" " * 5,
        skiprows=6,
        skipinitialspace=True,
        names=["year", "temp", "5-year-average"],
        engine="python",
    ).plot(
        ax=ax,
        x="year",
        y=["temp", "5-year-average"],
        legend=False,
        color=["dimgray", "black"],
    )

    ax.spines[["top", "right"]].set_visible(False)
    ax.hlines(
        0, 1880, 2022, linestyles="dashed", color="lightgray", alpha=0.5, zorder=0
    )
    ax.set_xlim(1880, 2022)
    return ax


fig, ax = plt.subplots()
plot_global_temp(ax)
ax.set_title(
    "Global Temperature Change, 1880-2022 (NASA/GISS)",
    loc="left",
    pad=10,
    fontsize=14,
    fontweight="bold",
    color="dimgray",
)
ax.set_ylabel("Change relative to 1951-80 average (°C)", fontsize=12, color="dimgray")
plt.show()
```

```python
def format_yticklabels_with_celsius(ax):
    yticks = ax.get_yticks()
    yticklabels = [f"{temp:.1f}°C" for temp in yticks]
    ax.set_yticks(yticks)
    ax.set_yticklabels(yticklabels)
    return ax
```

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(10, 10))
gspec = gridspec.GridSpec(nrows=12, ncols=12, figure=fig, hspace=20, wspace=6)

ax1 = fig.add_subplot(gspec[:5, :8])
plot_global_temp(ax1)
ax1.set_title(
    "Global Temperature Change, 1880-2022 (NASA/GISS)",
    loc="left",
    pad=10,
    fontsize=14,
    fontweight="bold",
    color="dimgray",
)
ax1.set_ylabel("Change relative to 1951-80 average (°C)", fontsize=12, color="dimgray")
format_yticklabels_with_celsius(ax1)

ax2 = fig.add_subplot(gspec[5:10, :8])
plot_global_temp(ax2)
ax2.set_ylim(-5, 5)
ax2.text(
    1885,
    5,
    "Misleading chart (1)",
    fontsize=12,
    color="red",
    ha="left",
    va="top",
    fontweight="bold",
)
format_yticklabels_with_celsius(ax2)

ax3 = fig.add_subplot(gspec[10:, :])
plot_global_temp(ax3)
ax3.text(
    1882,
    1.1,
    "Misleading chart (2)",
    fontsize=12,
    color="red",
    ha="left",
    va="top",
    fontweight="bold",
)
format_yticklabels_with_celsius(ax3)

ax4 = fig.add_subplot(gspec[:10, 8:])
plot_global_temp(ax4)
ax4.text(
    1885,
    1.1,
    "Misleading chart (3)",
    fontsize=12,
    color="red",
    ha="left",
    va="top",
    fontweight="bold",
)
format_yticklabels_with_celsius(ax4)

plt.show()
```
