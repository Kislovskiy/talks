```python
fig, ax = plt.subplots()
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
ax.plot(x, np.sin(x))
ax.set_title("sin")

ax.set_yticks([-1, 1])
ax.set_yticklabels(["-1", "+1"])

ax.set_xticks([-np.pi, -np.pi / 2, np.pi / 2, np.pi])
ax.set_xticklabels(["π-", "π-/2", "π+/2", "π+"])

plt.show()
```

```python
fig, ax = plt.subplots()
x = np.linspace(-np.pi, np.pi, 257, endpoint=True)
ax.plot(
    x,
    np.sin(x),
    linewidth=2,
    marker="o",
    markevery=(0, 64),
    markersize=8,
    markerfacecolor="white",
)

ax.set_yticks([-1, 1])
ax.set_yticklabels(["-1", "+1"])

ax.set_xticks([-np.pi, -np.pi / 2, np.pi / 2, np.pi])
ax.set_xticklabels(["-π", "-π/2", "+π/2", "+π"])

ax.spines[["bottom", "left"]].set_position(("data", 0))
ax.spines[["top", "right"]].set_visible(False)

ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

t = 2 * np.pi / 3

ax.annotate(
    r"$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$",
    xy=(t, np.sin(t)),
    xycoords="data",
    xytext=(+10, +30),
    textcoords="offset points",
    fontsize=16,
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"),
)

ax.plot([t, t], [0, np.sin(t)], color="grey", linewidth=1.5, linestyle="--")

fig.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patheffects import Stroke, Normal

fig = plt.figure(figsize=(8, 5))
ax = fig.add_axes([0, 0, 1, 1], frameon=False)
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.tick_params(
    axis="both", which="both", bottom=False, top=False, left=False, right=False
)
# family = "Pacifico"
size = 100
cmap = plt.cm.twilight
text = "Static"

for x in np.linspace(0, 1, 10):
    lw, color = x * 225, cmap(1 - x)
    t = ax.text(
        0.5,
        0.45,
        text,
        size=size,
        color="none",
        weight="bold",
        va="center",
        ha="center",
        # family=family,
        zorder=-lw,
    )
    t.set_path_effects([Stroke(linewidth=lw + 1, foreground="black")])
    t = ax.text(
        0.5,
        0.45,
        text,
        size=size,
        color="black",
        weight="bold",
        va="center",
        ha="center",
        # family=family,
        zorder=-lw + 1,
    )
    t.set_path_effects([Stroke(linewidth=lw, foreground=color)])

plt.show()
fig.savefig("../../results/static.png", dpi=300)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patheffects import Stroke, Normal

fig = plt.figure(figsize=(12, 5))
ax = fig.add_axes([0, 0, 1, 1], frameon=False)
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.tick_params(
    axis="both", which="both", bottom=False, top=False, left=False, right=False
)
# family = "Pacifico"
size = 100
cmap = plt.cm.terrain
text = "Dynamic"

for x in np.linspace(0, 1, 10):
    lw, color = x * 225, cmap(1 - x)
    t = ax.text(
        0.5,
        0.45,
        text,
        size=size,
        color="none",
        weight="bold",
        va="center",
        ha="center",
        # family=family,
        zorder=-lw,
    )
    t.set_path_effects([Stroke(linewidth=lw + 1, foreground="black")])
    t = ax.text(
        0.5,
        0.45,
        text,
        size=size,
        color="black",
        weight="bold",
        va="center",
        ha="center",
        # family=family,
        zorder=-lw + 1,
    )
    t.set_path_effects([Stroke(linewidth=lw, foreground=color)])

plt.show()
fig.savefig("../../results/dynamic.pdf", dpi=300)
```

```python
data = [10, 20, 30, 40, 50]
cumulative_sum = np.cumsum(data)

fig, ax = plt.subplots()
ax.plot(data, cumulative_sum, linewidth=3, marker="o", markersize=6, markerfacecolor="white")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_ylim(0, 160)
ax.spines[["top", "right"]].set_visible(False)
plt.show()
```

```python
data = [10, 20, 30, 40, 50]
transformed_data = np.cumsum(np.power(data, 2))

fig, ax = plt.subplots()
ax.plot(data, transformed_data, linewidth=3, marker="o", markersize=6, markerfacecolor="white")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_ylim(0, 160)
ax.spines[["top", "right"]].set_visible(False)
plt.show()
```

```python
def plot_something(axis, x, y, **kwargs):
    plt.xkcd()
    axis.plot(
        x,
        y,
        linewidth=3,
        marker="o",
        markersize=6,
        markerfacecolor="white",
        **kwargs
    )
    axis.set_xlabel("x")
    axis.set_ylabel("y")
    axis.spines[["top", "right"]].set_visible(False)
    return axis


data = [10, 20, 30, 40, 50]
fig, ax = plt.subplots(ncols=2, figsize=(12, 5))
    
for i, color in enumerate([None, "black"]):
    plot_something(axis=ax[i], x=data, y=np.cumsum(data), color=color)
    plot_something(axis=ax[i], x=data, y=np.cumsum(np.power(data, 2)), color=color)

    x = ax[i].lines[0].get_xdata()
    ax[i].fill_between(
        x=x,
        y1=ax[i].lines[0].get_ydata(),
        y2=ax[i].lines[1].get_ydata(),
        where=(x >= 40) & (x <= 50),
        alpha=0.3,
        color=color,
    )

plt.show()
```
